```python
#tkinter part 1
import tkinter as tk
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

window = tk.Tk()
window.title('Resistors In Parallel')
window.geometry('600x600')

#----------------------------------------------------------------------------------------

# Values of Voltage in (V) & Current IN (A):
VI = [(0.1 , 2e-3) , (0.2 , 4e-3) , (0.3 , 6.6e-3) , (0.4 , 8e-3) , (0.5 , 9.85e-3) ] # function

# Experimental Values of Resistance, Using Ohm's law:
def exp_R():
    x=0
    num_of_v = int(input('Please Enter the number of voltage values that you will measure: '))
    Rs=[]
    while x < num_of_v:
        V = float(input('Please Enter the voltage value: '))
        I = float(input('Please Enter the current value: '))
        A = [(V , I)]

        for v , i in A: #funtion
            #experimental_R = []
            Exp_R = v / i
            Rs.append(Exp_R)
            #experimental_R.append(Exp_R)
            print(f'The Experimental Resistance is: {round(Exp_R , 1)} Ω')
            print('-----------------------------------------')

        x+=1
    print(" Resistances = " , Rs)

    print(f'The Average Resistance is: {round((sum(Rs)/num_of_v))} Ω')
    lbl['text'] = str(round((sum(Rs)/num_of_v), ' Ω'))
    
    
#tkinter part 2
lbl = tk.Label(text=' ')

btn_Ravg = tk.Button(text='Calculate', command = exp_R)

lbl_Ravg = tk.Label(text='Average Resistance   = ')

btn_Ravg.grid(row=4 , column=6)
lbl.grid(row=3 , column=9)
lbl_Ravg.grid(row=3 , column=6)

#----------------------------------------------------------------------------------------


# Calculating the Equivalent Resistance
def Equivalent_Resistance():
    Eq_R =[]
    num_of_R = int(input('Please Enter the number of Resistors: '))
    z=0
    while z < num_of_R:
        value_of_R = float(input('Please Enter the value of Resistor: '))
        E = 1/value_of_R 
        Eq_R.append(E)
        z+=1
    print(f'The Equivalent Resistance is: {1/sum(Eq_R)} Ω')
    lblv['text'] = str(1/sum(Eq_R))
    
#tkinter part 3                     
lblv= tk.Label(text=' ')
btn_Req = tk.Button(text='Calculate', command = Equivalent_Resistance)
lbl_Req = tk.Label(text='Equivalent Resistance   = ')
lblv.grid(row=25 , column=11)
lbl_Req.grid(row=25 , column=6)
btn_Req.grid(row=27 , column=6)

#----------------------------------------------------------------------------------------

#tkinter part 4 
# Creating The IV Table:
def Make_IV_Table():
    Parallel_EC_data = pd.read_csv(r"C:\Users\أثير\Downloads\المصنف1.CSV")
    Parallel_EC_data
    lblt['text'] = Parallel_EC_data
    
lblt= tk.Label(text=' ')
btn_IVT = tk.Button(text='Show IV Table', command = Make_IV_Table)
lbl_IVT = tk.Label(text='IV Table      : ')

lbl_IVT.grid(row=34 , column=6)
btn_IVT.grid(row=37 , column=6)
lblt.grid(row=40 , column=6)

#----------------------------------------------------------------------------------------

#tkinter part 5
# Plotting the IV Graph:
def Plotting():
    
    ax = sns.lineplot(x= "I ( mA ) ", y= "  V (V)" , data = Parallel_EC_data , ci=None, markers=True)
    sns.set()
    
    lblp['text'] = ax
    
lblp= tk.Label(text=' ')
btn_plot = tk.Button(text='Show Graph', command = Plotting)
lbl_plot = tk.Label(text='IV Graph      : ')


lbl_plot.grid(row=44 , column=6)
btn_plot.grid(row=47 , column=6)
lblp.grid(row=50 , column=6)

#----------------------------------------------------------------------------------------

#tkinter part 6
# Calculating the slope:
import numpy as np

def Slope():
    x1 = float(input('Please Enter x1: '))
    x2 = float(input('Please Enter x2: '))
    y1 = float(input('Please Enter y1: '))
    y2 = float(input('Please Enter y2: '))

    x =[x1*1e-3,1e-3*x2]
    y =[y1,y2]
    slope, intercept = np.polyfit(x,y,1)
    print(slope)
    lbls['text'] = str(slope)
    
lbls= tk.Label(text=' ')
btn_slope = tk.Button(text='Calculate', command = Slope)
lbl_slope = tk.Label(text='Slope      : ')


lbl_slope.grid(row=54 , column=6)
btn_slope.grid(row=57 , column=6)
lbls.grid(row=60 , column=6)


#----------------------------------------------------------------------------------------

#tkinter part 7
# Calculting Percentage Error:
def PE():
    print(abs((50 - 50) / 50) * 100)
    lblPE['text'] = str(abs((50 - 50) / 50) * 100)
    
lblPE = tk.Label(text=' ')
btn_PE = tk.Button(text='Calculate', command = PE)
lbl_PE = tk.Label(text='Percentage Error %      : ')


lbl_PE.grid(row=64 , column=6)
btn_PE.grid(row=67 , column=6)
lblPE.grid(row=64 , column=8)





    



window.mainloop()
```
